```python
import numpy as np
import matplotlib.pyplot as plt


data = np.loadtxt("outParallel.txt")
print(data.shape)

plt.figure(figsize=(10, 5))
plt.imshow(data, cmap='hot')
plt.xlabel('x')
plt.ylabel('t')
plt.colorbar(label='u(x, t)')
plt.title('  ')
plt.tight_layout()
plt.show()
```
